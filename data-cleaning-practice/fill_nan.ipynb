```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'names': ['suchona', 'promy', 'sejuti', 'farhana'],
                   'age': [24, 24, 25, np.nan],
                   'cgpa': [3.50, np.nan, np.nan, 3.80],
                   'city': ['Dhaka', 'Sylhet', np.nan, 'Chittagong']
                   })
df
```

```python
# We can replace NaN values with empty string or something
df.fillna('', inplace=True)
df
```

```python
# fill NaN values in a single col

df['city'] = df.city.fillna('none')
df
```

```python
df[['age', 'cgpa']] = df[['age', 'cgpa']].fillna('zeros')
df
```

```python
df['age'] = df['age'].replace(np.nan, 'Not Provided')
df
```
